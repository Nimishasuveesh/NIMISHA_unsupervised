```python
import pandas as pd
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
```

```python
# Load the dataset
data = pd.read_csv("Wine_clust .csv")
```

```python
# Assuming the relevant columns for clustering are selected in the dataset
X = data.iloc[:, 1:].values  # Assuming the clustering features start from the second column
```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # Assuming 3 clusters
kmeans.fit(X_scaled)
kmeans_labels = kmeans.labels_
```

```python
# Hierarchical clustering (Agglomerative)
agg_cluster = AgglomerativeClustering(n_clusters=3)
agg_cluster.fit(X_scaled)
agg_labels = agg_cluster.labels_
```

```python
# DBSCAN Clustering
neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(X_scaled)
distances, indices = neighbors_fit.kneighbors(X_scaled)
distances = sorted(distances[:, -1])
plt.plot(distances)
plt.xlabel('Data Points sorted by distance')
plt.ylabel('Epsilon')
plt.show()
```

```python
# Visualization for K-means clustering
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans_labels, cmap='viridis', marker='o')
plt.title('K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar()
plt.show()
```

```python
# Visualization for Hierarchical clustering
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=agg_labels, cmap='viridis', marker='o')
plt.title('Agglomerative Hierarchical Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar()
plt.show()
```
